```r
# create weighted directed network
g = read.graph("project4_data/q1_weighted.txt",format="ncol",directed=TRUE)
```

```r

frequency = sort(degree(g, mode='in'))
df = data.frame(degree=c(NA),freq=c(NA)) #initialize dataframe
df = df[-1, ] #remove NA

current_degree = frequency[1]
count = 1

for (i in 2:length(frequency)){
    if(frequency[i] != current_degree){ #compare current_degree with next degree
        df[nrow(df)+1, ] = c(current_degree,count)
        current_degree = frequency[i]
        count = 1
        
    } else {
        #keep counting if they're the same.
        count = count + 1
    }
}
# add final degree count
df[nrow(df)+1, ] = c(current_degree, count)
```

```r
plot(df, type='p', col="lightblue",
     main="In-degree Distribution of Actor/Actress Network (frequency)",
    xlab="In-degree",ylab="Frequency")

plot(degree.distribution(g, mode="in"),
     type="p", col="lightblue",
     main="In-degree Distribution of the Actor/Actress Network (probability)",
     xlab="In-degree",ylab="Probability")
```
