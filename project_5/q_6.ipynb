```r
# QUESTION 6: Create a weighted undirected movie network using equation 2. Plot the degree
# distribution of the movie network. Briefly comment on the degree distribution.
```

```r
library(igraph)
```

```r
weighted_movie_g = read_graph("project4_data/q6_movies_weighted.txt",format="ncol",directed = FALSE)
```

```r
unique_degrees = unique(sort(degree(weighted_movie_g)))
if (unique_degrees[1]!=0) {
    unique_degrees = seq(0,max(unique_degrees))
}
dd = degree_distribution(weighted_movie_g)

plot(unique_degrees, dd, col="lightblue",
     main="Degree distribution of the movie network", xlab="Degree", ylab="Relative Frequency Vertices")
```
